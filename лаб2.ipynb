```python
# Імпортуємо бібліотеки
import pandas as pd        # робота з таблицями
import numpy as np         # робота з пропусками
import matplotlib.pyplot as plt   # графіки

# Завантажуємо файл CSV
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')

# Перші 5 рядків
print(df.head())    # показати початок таблиці

# Розмір таблиці
print(df.shape)     # рядки і стовпці

# Кількість пропусків
print(df.isna().sum())   # перевірка пустих клітинок

# Замінюємо пусті значення на NaN
df = df.replace('', np.nan)   # пусто = NaN

# Числові стовпці переводимо в числа
df['video views'] = pd.to_numeric(df['video views'], errors='coerce')
df['uploads'] = pd.to_numeric(df['uploads'], errors='coerce')

# Заповнюємо пропуски середнім
df['video views'] = df['video views'].fillna(df['video views'].mean())
df['uploads'] = df['uploads'].fillna(df['uploads'].mean())

# Тип даних float
df['video views'] = df['video views'].astype(float)
df['uploads'] = df['uploads'].astype(float)

# Ще раз перевірка пропусків
print(df.isna().sum())

# Скільки унікальних країн
print(df['Country'].nunique())

# Гістограма переглядів
df['video views'].plot(kind='hist', bins=20)
plt.title('Розподіл переглядів')
plt.xlabel('Перегляди')
plt.show()

# Максимум, мінімум, середнє
print(df['video views'].max())    # максимум
print(df['video views'].min())    # мінімум
print(df['video views'].mean())   # середнє

# Країна з найбільшою кількістю каналів
print(df['Country'].value_counts().idxmax())

# Канал з найбільшим uploads
print(df.loc[df['uploads'].idxmax(), 'Youtuber'])

# Канал з найменшим uploads
print(df.loc[df['uploads'].idxmin(), 'Youtuber'])
```

```python


import pandas as pd                      # підключаємо бібліотеку pandas для роботи з таблицями

df = pd.read_csv("bestsellers with categories.csv")   # зчитуємо csv файл у таблицю DataFrame

print(df.head(10))                      # виводимо перші 10 рядків таблиці

print(df.shape)                         # показує кількість рядків і стовпців

print(df.shape[0])                      # показує кількість книг (рядків)

df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
# змінюємо назви всіх стовпців

print(df.columns)                       # виводимо нові назви стовпців

print(df.isnull().sum())                # рахуємо кількість пропусків у кожному стовпці

print(df.isnull().sum().sum())          # загальна кількість пропусків у таблиці

print(df['genre'].unique())             # показує всі унікальні жанри

print(df['price'].max())                # максимальна ціна книги

print(df['price'].min())                # мінімальна ціна книги

print(df['price'].mean())               # середня ціна книги

print(df['price'].median())             # медіанна ціна книги

print(df['user_rating'].max())          # найвищий рейтинг у датасеті

print(df[df['user_rating'] == df['user_rating'].max()].shape[0])
# кількість книг з найвищим рейтингом

print(df[df['reviews'] == df['reviews'].max()][['name', 'reviews']])
# книга з найбільшою кількістю відгуків

print(df[df['year'] == 2010].sort_values('price', ascending=False).head(1))
# найдорожча книга серед топ-50 у 2010 році

print(df[(df['year'] == 2012) & (df['genre'] == 'Fiction')].shape[0])
# кількість книг жанру Fiction у 2012 році

print(df[(df['user_rating'] == 4.9) & (df['year'].isin([2010, 2011]))].shape[0])
# кількість книг з рейтингом 4.9 у 2010 і 2011 роках

print(df.groupby('genre')['price'].agg(['min', 'max']))
# мінімальна і максимальна ціна по жанрах

df.to_csv("Мартинюк_ЛР_2.csv", index=False)
# зберігаємо змінений файл у csv
```
