```python
# Лабораторна робота 3

import pandas as pd                           # підключаємо pandas для роботи з таблицями
import matplotlib.pyplot as plt              # підключаємо matplotlib для графіків

df = pd.read_csv('spotify-2023.csv', encoding='latin1')   # зчитуємо файл CSV у таблицю

print(df.head())                             # показуємо перші 5 рядків таблиці
print(df.columns)                            # показуємо назви стовпців



top10 = df.sort_values(by='streams', ascending=False).head(10)   # сортуємо по прослуховуваннях і беремо топ 10

plt.figure(figsize=(10,6))                  # розмір графіка
plt.barh(top10['track_name'], top10['streams'], color='skyblue') # горизонтальні стовпчики
plt.title('ТОП 10 пісень Spotify 2023')     # заголовок
plt.xlabel('Кількість прослуховувань')      # підпис осі X
plt.ylabel('Назва треку')                   # підпис осі Y
plt.gca().invert_yaxis()                    # найбільше значення зверху
plt.show()                                  # показати графік



genre_data = df['artist(s)_name'].value_counts().head(5)   # беремо 5 найпопулярніших артистів

plt.figure(figsize=(8,8))                  # розмір графіка
plt.pie(genre_data, labels=genre_data.index, autopct='%1.1f%%') # кругова діаграма
plt.title('Розподіл популярних артистів')   # заголовок
plt.show()                                  # показати графік



song = df[df['track_name'] == df['track_name'][0]]   # вибираємо перший трек

months = ['Січ','Лют','Бер','Кві','Тра','Чер','Лип','Сер','Вер','Жов','Лис','Гру'] # місяці

streams = [100,120,140,160,180,210,240,260,300,340,370,400]   # умовні дані популярності

plt.figure(figsize=(10,5))                 # розмір графіка
plt.plot(months, streams, marker='o', color='green') # лінійний графік
plt.title('Динаміка популярності треку')   # заголовок
plt.xlabel('Місяць')                       # підпис осі X
plt.ylabel('Прослуховування')              # підпис осі Y
plt.grid(True)                             # сітка
plt.show()                                 # показати графік



artists = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10) # топ 10 артистів

plt.figure(figsize=(10,6))                # розмір графіка
plt.bar(artists.index, artists.values, color='orange') # вертикальні стовпчики
plt.title('ТОП 10 артистів Spotify')      # заголовок
plt.xlabel('Артист')                      # підпис осі X
plt.ylabel('Прослуховування')             # підпис осі Y
plt.xticks(rotation=45)                   # повертаємо назви артистів
plt.show()                                # показати графік



three = df[['track_name','streams']].head(3)   # беремо перші 3 треки

plt.figure(figsize=(10,5))                # розмір графіка
plt.barh(three['track_name'], three['streams'], color='purple') # горизонтальний графік
plt.title('Порівняння трьох треків')      # заголовок
plt.xlabel('Прослуховування')             # підпис осі X
plt.ylabel('Трек')                        # підпис осі Y
plt.show()                                # показати графік
```
